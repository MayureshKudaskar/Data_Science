# Creating a Pivot table

```python

user_df=book.pivot_table(index='userID',columns='title',values='rating').reset_index(drop=True)
userID=user_df.index
title=user_df.columns
user_df
```

# Creating index from UserID.

```python

user_df.index=book.userID.unique()
```

#Filling the NA values with zero


```python
user_df.fillna(0, inplace =True)
user_df
```

```python

# Import libraries to find similarities 
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
# FInding Similarities
user_sim=1-pairwise_distances(user_df,metric='cosine')
user_sim
```

```python

user_sim_df=pd.DataFrame(user_sim)
user_sim_df.index=book.userID.unique()
user_sim_df.columns=book.userID.unique()
user_sim_df
```

```python
# Filling the diagonal cells with zero 
np.fill_diagonal(user_sim,0)
user_sim_df
```

# Most similar users

```python

user_sim_df.idxmax(axis=1)[0:20]
```
