```python
cos = pd.DataFrame(index=['Error Free','Defective'],
                  data={'Phillippines':[271,29],'Indonesia':[267,33],'Malta':[269,31],'India':[280,20]})
```

```python
cos
```

```python
cos.describe()
```

```python

chi2_stat, p_val, dof, ex =stats.chi2_contingency(cos) 

print("===Chi2 Stat===")
print(chi2_stat)
print("\n")
print("===Degrees of Freedom===")
print(dof)
print("\n")
print("===P-Value===")
print(p_val)
print("\n")
print("===Contingency Table===")
print(ex)
```


#Since p-value (0.277)> alpha (0.05), hence we can't reject the null hypothesis
#Conclusion: Percentages of Defective across all the centres is same
