```python
FL1 = pd.DataFrame(index=['Female','Male'],
                  data={'Weekdays':[287,113],'Weekend':[233,167]})
```

```python
FL1
```

```python
from scipy.stats import chi2_contingency 
```

```python
chi2_stat, p_val, dof, ex =stats.chi2_contingency(FL1) 

print("===Chi2 Stat===")
print(chi2_stat)
print("\n")
print("===Degrees of Freedom===")
print(dof)
print("\n")
print("===P-Value===")
print(p_val)
print("\n")
print("===Contingency Table===")
print(ex)
```

#Since p-value (0.8534e-05)<alpha (0.05), hence we have to reject the null hypothesis
#Proportions of Male and Female are not same.
